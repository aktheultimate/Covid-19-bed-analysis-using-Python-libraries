```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Data source https://www.kaggle.com/ikiulian/global-hospital-beds-capacity-for-covid19
```

```python
dfi = pd.read_csv("hospitalbedsindiav1.csv")#importing  hospital bed capacity for covid-19 data for India as dfi
dfu = pd.read_csv("hospitalbedsus.csv")#importing  hospital bed capacity for covid-19 data for USA as dfu
#Dropping duplicates to be sure that data is cleaned
dfi.drop_duplicates(inplace=True)
dfu.drop_duplicates(inplace=True)
```

```python
#Let's drop any data that has state also as "US" because US is a country, not a state
dfu.drop(dfu[dfu.state=="US"].index)
dfu
```

```python
#Analysis between source year for beds and population data of the two countries side by side to see which of it is more recent
print("Indian beds (year) data")
print(dfi.year.value_counts(),"\n")
print("American beds (year) data")
print(dfu.year.value_counts())
yearu = dfu.year.unique()
yearu = np.sort(yearu)
yeari = dfi.year.unique()
yeari = np.sort(yeari)
plt.bar(yearu-0.2,dfu.year.value_counts(),width=0.4,label="American beds")
plt.bar(yeari+0.2,dfi.year.value_counts(),width=0.4,label="Indian beds")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Number of beds")
plt.title("Number of beds added per year")
plt.show()
```

```python
print("We can see that the number of beds in USA are being updated in 2020 too while as far as Indian data is concerned, the number of beds haven't been updated from 2018. This might also be due to the fact that our USA based data is much more vast and has more entries \'5712' ""than Indian data")
```

```python
print("It is also worth noticing that the maximum number of beds were put up in 2018 in both the countries")
```

```python
#Maximum number of bed present in which state and county in each country and how much with their population
print("In India, maximum number of beds are present in:\n",dfi.loc[dfi.beds==dfi.beds.max(),['state','county','population','beds']])
print("In USA, maximum number of beds are present in:\n",dfu.loc[dfu.beds==dfu.beds.max(),['state','county','population','beds']])
"""In India, the state with maximum beds is Tamil Nadu"""
"""In USA, the state and county with maximum beds is Kansas, Pawnee"""
```

```python
#Now, let's see where the maximum population resides in bith the countries
print("In India, maximum population resides in:\n",dfi.loc[dfi.population==dfi.population.max(),['state','county','population','beds']])
print("In USA, maximum population resides in:\n",dfu.loc[dfu.population==dfu.population.max(),['state','county','population','beds']])
```

```python
"""As we can see that in both the countries the maximum population present isn't having the maximum beds where the population is maximum"""
```

```python
ip,ib = dfi[["population","beds"]].sum(axis=0) #Total number of beds i(in 1000) and population in India
```

```python
up,ub = dfu[["population","beds"]].sum(axis=0) #Total number of beds (in 1000) and population in USA
```

```python
#Visualization of total number of beds present
country = np.array([0,1])
pop = np.array([ip,up])
beds = np.array([ib,ub])
plt.xlabel("Countries")
plt.ylabel("Number of population and beds(per thousand)")
plt.title("Population vs Beds(per thousand)")
plt.xticks(country,["India","USA"])
plt.bar(country-0.2,pop/1000,width=0.4,label="Population")
plt.bar(country+0.2,beds,width=0.4,label="Beds")
plt.legend()
plt.show()
```

```python
#As we can see that there is not enough beds for the whole population in both the countries
#India has more than twice of the population of USA but even lesser beds than USA, which is a matter of concern
```

```python
#Analysis of latitude wise population distribution
#India
plt.xlabel("Latitude")
plt.ylabel("Number of population(per thousand)")
plt.title("Population distribution latitude wise")
plt.bar(dfi.lat,(dfi.population/1000),width=0.4,label="India")
plt.legend()
plt.show()
#Analysis of latitude wise bed distribution
#India
plt.xlabel("Latitude")
plt.ylabel("Number of beds(per thousand)")
plt.title("Beds distribution latitude wise")
plt.bar(dfi.lat,(dfi.beds),width=0.4,label="India")
plt.legend()
plt.show()
```
